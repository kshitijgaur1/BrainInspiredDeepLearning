```python
import numpy as np
import matplotlib.pyplot as plt
```

# LEAKY Intergrate and Fire Model

```python
R = 10
tau = 10
V0 = -70
Vth = -55
V_spike = 20
V_reset = -75
step_size = 0.1
```

```python
t = np.arange(0, 50, step_size)
I = 2*np.ones(len(t))
I[int(len(t)/2):] = 0
Vm = np.zeros(len(t))
Vm[0] = V0
```

```python
iter = 1
while iter < len(t):
    if Vm[iter-1] >= Vth:
        Vm[iter] = V_spike
        Vm[iter+1] = V_reset
        iter = iter+2
    else:
        Vm[iter] = Vm[iter-1] + step_size*(((V0-Vm[iter-1])/tau) + I[iter-1]*R/tau)
        iter = iter+1
```

```python
plt.plot(t, Vm)
plt.plot(t, Vth*np.ones(len(t)), '--')
plt.xlabel("time ms")
plt.ylabel("Membrane Potential (mV)")
```
